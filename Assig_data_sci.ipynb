```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('/content/iris.xls')
```

```python
data
```

```python
# columns in the dataset
data.columns
```

```python
# Mean of each column of the dataset.
data[['SL', 'SW', 'PL', 'PW']].mean()
```

```python
# null values present in the dataset
data.isna().sum()
```

```python
# VISAUALIZATIONS
data.plot(kind = 'line')
```

```python
data.plot(kind='box', subplots=True, layout=(2,2), sharex=False, sharey=False)
```

```python
data.groupby('Classification').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
figsize = (12, 1.2 * len(data['Classification'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(data, x='SL', y='Classification', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```
